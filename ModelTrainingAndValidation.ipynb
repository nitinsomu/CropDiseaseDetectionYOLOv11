```python
import os

def count_files_in_folder(folder_path):
    total_files = 0
    
    for item in os.listdir(folder_path):
        item_path = os.path.join(folder_path, item)
        if os.path.isfile(item_path):
            total_files += 1
        elif os.path.isdir(item_path):
            total_files += count_files_in_folder(item_path)
    
    return total_files
```

```python
import numpy as np

TP = np.array(TP)
FP = np.array(FP)
FN = np.array(FN)

# Precision, Recall, F1 for each class
precision_per_class = TP / (TP + FP + 1e-9)
recall_per_class = TP / (TP + FN + 1e-9)
f1_per_class = 2 * (precision_per_class * recall_per_class) / (precision_per_class + recall_per_class + 1e-9)

# Macro averages
macro_precision = precision_per_class.mean()
macro_recall = recall_per_class.mean()
macro_f1 = f1_per_class.mean()

# Micro averages
total_TP = TP.sum()
total_FP = FP.sum()
total_FN = FN.sum()

micro_precision = total_TP / (total_TP + total_FP)
micro_recall = total_TP / (total_TP + total_FN)
micro_f1 = 2 * (micro_precision * micro_recall) / (micro_precision + micro_recall)

macro_precision, macro_recall, macro_f1, micro_precision, micro_recall, micro_f1
```

```python
# Per-class confusion matrix values: (TP, FP, FN)
class_stats = [
    (252, 0, 0), (249, 0, 0), (220, 0, 0), (251, 0, 0), (227, 0, 1),
    (211, 0, 0), (228, 0, 0), (199, 6, 1), (239, 0, 1), (238, 1, 5),
    (233, 0, 0), (236, 0, 0), (240, 0, 0), (215, 0, 0), (212, 0, 0),
    (252, 0, 0), (230, 0, 0), (216, 0, 0), (239, 0, 0), (249, 0, 0),
    (243, 0, 0), (243, 0, 0), (227, 1, 0), (223, 0, 0), (253, 0, 0),
    (217, 0, 0), (222, 0, 0), (228, 0, 0), (212, 1, 1), (239, 1, 1),
    (231, 1, 0), (235, 0, 0), (218, 0, 1), (218, 0, 1), (227, 2, 0),
    (244, 1, 1), (224, 0, 0), (241, 0, 1)
]
```

```python
def calculate_map(class_stats):
    ap_list = []
    for tp, fp, fn in class_stats:
        if (tp + fp) == 0 or (tp + fn) == 0:
            ap = 0  # Handle edge cases
        else:
            precision = tp / (tp + fp)
            recall = tp / (tp + fn)
            ap = precision  # AP ≈ precision if one threshold
        ap_list.append(ap)
    map_score = sum(ap_list) / len(ap_list)
    return map_score, ap_list
```

```python
map_score, ap_list = calculate_map(class_stats)
```

```python
print(f"Mean Average Precision (mAP): {map_score:.4f}")
```
